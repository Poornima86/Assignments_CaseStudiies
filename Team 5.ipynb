# Data Cleaning

```python

data_comp.km_driven = data_comp.km_driven.replace({',': '', 'kms': ''}, regex=True).astype(int)

data_comp.mileage = data_comp.mileage.str.extract('([0-9]+[.][0-9]+)').astype(float)

data_comp.engine = data_comp.engine.str.extract('([0-9]+)').astype(dtype=float)

data_comp.max_power = data_comp.max_power.str.extract('([0-9]+[.]?[0-9]*)').astype(dtype=float)

data_comp.seats = data_comp.seats.str.extract('([0-9]+)').astype(dtype=int, errors='ignore')
data_comp.head()
```

```python
for i in range(data_comp.shape[0]):
     try:
        price = float(data_comp['selling_price'][i].split(' ')[0])
        digit = data_comp['selling_price'][i].split(' ')[1]
        if digit == 'Lakh*':
            price = price * 100000
            data_comp['selling_price'][i] = price
        elif digit == 'Cr*':
            price = price * 10000000
            data_comp['selling_price'][i] = price
     except:
        price = data_comp['selling_price'][i][:-1]
        price = price.replace(',', '')
        data_comp['selling_price'][i] = float(price)
```

```python
data_comp=data_comp.convert_dtypes()
data_comp.head()
```

```python
data_comp.new_price = data_comp.new_price.str.extract('([0-9]+[.][0-9]+-[0-9]+[.][0-9]+)').astype(dtype=int, errors='ignore')
data_comp.head()
```

```python
data_comp['min_new_price'] = data_comp.new_price.str.extract('([0-9]+[.][0-9]+)').astype(dtype=float)
data_comp['max_new_price'] = data_comp.new_price.str.extract('\-(.*)').astype(dtype=float)
data_comp.head()
```

```python
data_comp['company'] = data_comp['full_name'].str.split(' ', expand=True)[0]
```

```python
data_comp.drop(columns=['new_price','full_name','owner_type','min_new_price','max_new_price'], axis=1, inplace=True)
data_comp.head()
```

```python
def removeOutliers(data, col):
    Q3 = np.quantile(data[col], 0.75)
    Q1 = np.quantile(data[col], 0.25)
    IQR = Q3 - Q1
      
    print("IQR value for column %s is: %s" % (col, IQR))
    global outlier_free_list
    global filtered_data
      
    lower_range = Q1 - 1.5 * IQR
    upper_range = Q3 + 1.5 * IQR
    outlier_free_list = [x for x in data[col] if (
        (x > lower_range) & (x < upper_range))]
    filtered_data = data.loc[data[col].isin(outlier_free_list)]

out_columns = df[['km_driven','vehicle_age','mileage','engine','max_power','seats','selling_price','min_cost_price','max_cost_price']]  
for i in out_columns:
    removeOutliers(df, i)
```
